```python
import numpy as np

A1 = np.matrix('1 2; 3 4')
print(A1)
```

```python
B1 = np.matrix('3 0 -1; 1 -1 2')
print(B1)
```

```python
C1 = A1*B1
print(C1)
```

```python
A2 = np.matrix('1 2; 3 4')
print(A2)
B2 = np.matrix('3 0; 1 -1')
print(B2)
C2 = A2*B2
print(C2)
```

```python
D2 = B2*A2
print(D2)
```

```python
Dinv = np.linalg.inv(D2)
print(Dinv)
```

```python
X3 = np.matrix('2; -1; 3')
A3 = np.matrix('2 1 -1; 1 -2 1; -1 2 3')
B3 = A3*X3
print('A3=',A3)
print('X3=',X3)
print('B3=',B3)
A3inv = np.linalg.inv(A3)
print('A3inv=',A3inv)
X4 = A3inv*B3
print('X4=',X4)
```
