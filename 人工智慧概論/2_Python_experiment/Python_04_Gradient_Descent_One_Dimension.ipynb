# Gradient descent: f(x) = x^^2 + 20*sin(x)

### (1) import libraries

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

### (2) set up x and y as array

```python
x = np.zeros(1600)
y = np.zeros(1600)
print(x.dtype, x.size, y.dtype, y.size)
```

### (3) f(x) = x^^2

```python
for i in range(0,1600): 
    x[i]=i/100 - 8
    y[i]=math.pow(x[i],2)
plt.plot(x,y)
plt.ylim([-40,60])
plt.grid()
```

### (4) f(x) = 20*sin(x)

```python
for i in range(0,1600): 
    x[i]=i/100 - 8
    y[i]=20*math.sin(x[i])
plt.plot(x,y)
plt.ylim([-40,60])
plt.grid()
```

### (5) f(x) = x^^2 + 20*sin(x)

```python
for i in range(0,1600):
     x[i]= i/100 - 8
     y[i]= math.pow(x[i],2)+20*math.sin(x[i])
  
plt.plot(x,y,color='red')
plt.ylim([-40,60])  
plt.grid()
```

### (6) df/dx = f'(x) = 2*x + 20*cos(x)

```python
df = np.zeros(1600)

for i in range(0,1600):
     x[i]= i/100 - 8
     y[i]= math.pow(x[i],2)+20*math.sin(x[i])
     df[i] = 2*(x[i])+20*math.cos(x[i])

plt.plot(x,y,color='red')
plt.plot(x,df,color='black')
plt.ylim([-40,60])
plt.grid()
plt.show()
```

### (7) Set up the initial value, start at cur_x = -4 or 1; rate = 0.01, 0.05, 0.1

```python
cur_x = -4 # The algorithm starts at x = -4
rate = 0.01 # Learning rate
precision = 0.001 #This tells us when to stop the algorithm
previous_step_size = 1 #
max_iters = 50 # maximum number of iterations
iters = 0 #iteration counter
df = lambda x: 2*x+20*math.cos(x) #Gradient of our function 

# ref: https://towardsdatascience.com/implement-gradient-descent-in-python-9b93ed7108d1
```

### (8) Do the iteration

```python
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x #Store current x value in prev_x
    cur_x = cur_x - rate * df(prev_x) #Grad descent
    previous_step_size = abs(cur_x - prev_x) #Change in x
    iters = iters+1 #iteration count
    print("Iteration",iters,"\nX value is",cur_x) #Print iterations
    
print("The local minimum occurs at", cur_x)

# ref: https://towardsdatascience.com/implement-gradient-descent-in-python-9b93ed7108d1
```
